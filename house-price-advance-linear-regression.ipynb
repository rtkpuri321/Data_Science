```python
import cupy as cp
import cudf as cd
import cuml as cm
```

```python
train=cd.read_csv("../input/../input/house-prices-advanced-regression-techniques/train.csv")
test=cd.read_csv("../input/../input/house-prices-advanced-regression-techniques/test.csv")
```

```python
train.shape,test.shape
```

```python
test.head()
```

```python
Y = train['SalePrice']
train.drop('SalePrice',axis=1,inplace=True)
```

```python
# Drop columns having more than 80% missing values
#to_drop = []
categorical_features = []
for col in train.keys():
    if train[col].dtype == 'object':
        categorical_features.append(col)
        
    #if train[col].isnull().sum()>=len(train)//1.25:
     #   to_drop.append(col)

#train = train.drop(to_drop,axis=1)
```

```python
for col in train.keys():
    if train[col].isnull().sum()!=0:
        if col in categorical_features:
            train[col].fillna(train[col].mode()[0],inplace=True)
        else:
            train[col].fillna(train[col].mean(),inplace=True)
    if test[col].isnull().sum()!=0:
        if col in categorical_features:
            test[col].fillna(test[col].mode()[0],inplace=True)
        else:
            test[col].fillna(test[col].mean(),inplace=True)
```

```python
test.isnull().any().sum()
```

```python
# encode categorical variables
le = cm.preprocessing.LabelEncoder()
for name in train.columns:
    if train[name].dtype == 'O':
        train[name] = train[name].astype(str)
        le.fit(train[name])
        train[name] = le.transform(train[name])
    
    if test[name].dtype == 'O':
        test[name] = test[name].astype(str)
        le.fit(test[name])
        test[name] = le.transform(test[name])
```

```python
train.head()
```

```python
train.isnull().any().sum()
```

```python
X=train
X.shape,test.shape
```

```python
Y.shape
```

```python
x_train,x_test,y_train,y_test=cm.train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
train.isnull().any().sum()
```

```python
y_train
```

**SVD**

```python
model = cm.LinearRegression(algorithm="svd")
trained = model.fit(x_train.astype('float32'),y_train.astype('float64'))
Y_predict = trained.predict(x_test)

mse = cm.metrics.regression.mean_squared_error(y_test.astype('int64'),Y_predict.astype('int64'))
mae = cm.metrics.regression.mean_absolute_error(y_test.astype('int64'),Y_predict.astype('int64'))
r2 = cm.metrics.regression.r2_score(y_test.astype('float64'),Y_predict.astype('float64'))
print("Algorithm SVD:")
print("R2 score:",str(r2))
print("MSE:",str(mse))
print("MAE:",str(mae))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(cp.asarray(Y_predict).get(), cp.asarray(y_test).get(), alpha = 0.7, color = 'b')
plt.xlabel('Predicted Price')
plt.ylabel('Actual Price')
plt.title('Linear Regression Model svd')
```
